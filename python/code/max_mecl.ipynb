```python
def max_mecl(ncount, SFR, FeH, Mecl_max, Kecl):
    Mecl = [0.0] * ncount
    Beta = -0.106*np.log10(SFR) + 2.
    Mecl[0] = 5.0
    M_t = 0.0
    Mtot = SFR*10.**7.0  #Mtot=SFR*dt
    Mecl_u = 10.**9.0
    Mecl_min = 5.0
    for f = range(1,900):
        M_t = 0.
        Mecl[f] = Mecl_min + 10**(0.01*f)
        if Beta == 2.:
            M_t = (np.log10(Mecl[f]) - np.log10(Mecl_min)) / (-1.0 * Mecl[f]**(-1.0) + Mecl_u**(-1.0))
        else:
            M_t = ((1.0-Beta)/(2.0-Beta))*(Mecl[f]**(2.0-Beta) - Mecl_min**(2.0-Beta))*
                                                                            ((Mecl_u**(1.0-Beta) - Mecl[f]**(1.0-Beta))**(-1.0))
        if M_t >= Mtot:
            Mecl_max = Mecl[f]
            break
        print(Mecl_max,Kecl,SFR)
        print(f,M_t,Mtot,Mecl(f),Mecl_max)
        
    if M_t < Mtot:
        Mecl_max = 1.0e9

    Kecl = Mtot*(-Beta + 1.0) / (Mecl_max ** (-Beta + 1.0) - Mecl_min ** (-Beta + 1.0))
    print("inja", Kecl, Mecl_max)
    print(Mecl_max, Kecl, SFR)
```

```python
def max_mstar(Mecl1,ncount,SFR,FeH,M_L,M_tu1,M_tu2,M_U,alpha1,alpha2,m_max2):
    m = [0] * ncount
    dm = [0] * ncount
    logM = [0] * ncount
     
    logM[0] = -1.08
    m[0] = 10.0 ** (logM[0])           
    IX1 = 0.99 * (0.61 * np.log10(Mecl1) + 2.85 - 6.0) - 0.14 * FeH
    if IX1 >= -0.87:
        alpha3 = 1.0 * (1.94 - 0.41 * IX1)
    elif IX1 < -0.87:
        alpha3=2.3
        
    M_C=0.0     
    for j in range(1,400):
        logM[j] = -1.08 + 0.01 * (j-1)  
        m[j] = 10.0 ** logM[j]
        dm[j] = m[j] - m[j-1]
        M_C=0.0                  

        if m[j] >= 0.08 and m[j] < 0.5:
            K1 = 1.0 / ((M_tu1 ** (1.0 - alpha1) - m[j] ** (1.0 - alpha1)) / (1.0 - alpha1) + (M_tu2 ** (1.0 - alpha2)\
            - M_tu1 ** (1.0 - alpha2)) * (M_tu1 ** (alpha2 - alpha1)) / (1.0 - alpha2) + (M_U ** (1.0 - alpha3)\
            - M_tu2 ** (1.0 - alpha3)) * (M_tu1 ** (alpha2 - alpha1)) * (M_tu2 ** (alpha3 - alpha2)) / (1.0 - alpha3))
                   
            M_C = K1 * (m[j] ** (2.0 - alpha1) - M_L ** (2.0 - alpha1)) / (2.0 - alpha1)
            
        if (m[j] >= 0.5 and m[j] < 1.0):
            K1 = 1.0 / ((M_tu2 ** (1.0 - alpha2) - m[j] ** (1.0 - alpha2)) * (M_tu1 ** (alpha2 - alpha1)) / (1.0 - alpha2)\
            + (M_U ** (1.0 - alpha3) - M_tu2 ** (1.0 - alpha3)) * (M_tu1 ** (alpha2 - alpha1))\
            * (M_tu2 ** (alpha3 - alpha2)) / (1.0 - alpha3))
            
            M_C = K1 * ((M_tu1 ** (2.0 - alpha1) - M_L ** (2.0 - alpha1)) / (2.0 - alpha1) + (m[j] ** (2.0 - alpha2)\
            - M_tu1 ** (2.0 - alpha2)) * (M_tu1 ** (alpha2 - alpha1)) / (2.0 - alpha2))
            
        if m[j] >= 1.0 and m[j] < 150.0:
            K1 = (M_tu1 ** (alpha1 - alpha2)) * (M_tu2 ** (alpha2 - alpha3)) * (1.0 - alpha3)\
            / (M_U ** (1.0 - alpha3) - m[j] ** (1.0 - alpha3))
                   
            M_C = K1 * ((M_tu1 ** (2.0 - alpha1) - M_L ** (2.0 - alpha1)) / (2.0 - alpha1) + (M_tu2 ** (2.0 - alpha2)\
            - M_tu1 ** (2.0 - alpha2)) * (M_tu1 ** (alpha2 - alpha1)) / (2.0 - alpha2) + (M[j] ** (2.0 - alpha3)\
            - M_tu2 ** (2.0 - alpha3)) * (M_tu2 ** (alpha3 - alpha2)) * (M_tu1 ** (alpha2 - alpha1)) / (2.0 - alpha3))
      
        if M_C >= Mecl1:
            m_max2=m[j]
            break
        print(m_max1)


    if M_C < Mecl1:
        m_max2 = 150.0
  
    print(m_max1)
    print(Mecl1,m_max2)
```
